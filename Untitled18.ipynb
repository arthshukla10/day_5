```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Arth Shukhla\Downloads\train.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in ['Survived', 'Pclass', 'Sex', 'Embarked']:
    sns.countplot(x=col, data=df)
    plt.title(f"Count Plot for {col}")
    plt.show()
```

```python
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Fare'].hist(bins=20)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Bivariate Analysis
# Survival Rate by Gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival Rate by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
# Age vs Survived
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.pairplot(df, hue='Survived', vars=['Pclass', 'Age', 'Fare'])
plt.show()
```

Summary of EDA Findings:
- Females had a much higher survival rate compared to males.
- Passengers in 1st class had a better chance of survival than those in 2nd or 3rd class.
- Younger passengers tended to survive more often than older ones.
- Higher fare was slightly associated with higher survival.
- Strong correlation between Pclass and Fare.
